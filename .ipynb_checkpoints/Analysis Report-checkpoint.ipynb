## __b. Checking for duplicate values__
Next, we check for duplicate entries to ensure each product is uniquely represented.

```python
duplicates = data.duplicated().sum()
print("Duplicates:", duplicates)
if duplicates == 0:
    print("No duplicate data found.")
else:
    print("Duplicate data found.")
```

# __3. Exploratory Data Analysis with Data Visualization__ 

## __a. Basic statistics__

```python
print("Basic statistics:\n", data.describe())
```

## __b. Distribution of Product Prices__
Visualize the distribution of product prices to understand the price range and frequency.

```python
plt.figure(figsize=(12,6))
plt.hist(data['price'], bins=30, color="violet", edgecolor='k', alpha=0.8)
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

* The price range appears to be from 0 to around 500.
* There are more products in the lower price range than in the higher price range.
* The distribution is right-skewed, meaning there are more products towards the lower end of the price range.

## __c. Average Review Score per Category__
Analyze the average review scores for each product category.

```python
category_review_count = data.groupby('category')['review_count'].sum()
category_review_count
```

```python
plt.figure(figsize=(10,5))
plt.bar(category_review_count.index, category_review_count.values, color='lightgreen', edgecolor='k')
plt.title('Average Review Score per Category')
plt.xlabel('Category')
plt.ylabel('Total Review Count')
plt.xticks(rotation=45)
plt.show()
```

* Sports has higher review counts indicating that it is more popular products.

## __d. Sales Distribution by Category__
Evaluate the sales distribution across different product categories.

```python
sales_columns = data.loc[:, 'sales_month_1':'sales_month_12']
```

```python
product_category = data.groupby("category")[sales_columns.columns].sum()
total_sales_by_category = product_category.sum(axis=1)
total_sales_by_category
```

```python
plt.figure(figsize=(10,5))
plt.bar(total_sales_by_category.index, total_sales_by_category.values, color="lightgreen", edgecolor='k')
plt.title('Sales Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

* Books has the highest total sales which indicates strong demand among customers.
* Home & Kitchen has low sales, which indicates that it needs further investigation for potential improvements in product offerings, marketing, or customer engagement.

## __e. Monthly Sales__
Analyze the sales trend over the year to identify any patterns or seasonal effects.

```python
monthly_totals = data.loc[:, 'sales_month_1':'sales_month_12'].sum()
monthly_totals
```

```python
plt.figure(figsize=(12, 6))
plt.plot(monthly_totals.index, monthly_totals.values, marker='o', color="green")
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(ticks=range(len(monthly_totals.index)), labels=monthly_totals.index, rotation=45)
plt.show()
```

* Highest sales was done on Sales-Month-10 while lowest sales was done on Sales-Month-5.

## __f. Price vs. Review Score__
Investigate the relationship between price and review scores.

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['price'], data['review_score'], color="purple")
plt.title('Price vs. Review Score')
plt.xlabel('Price')
plt.ylabel('Review Score')
plt.show()
```

* There is no strong correlation between price and review score i.e., both high-priced and low-priced products can have varied review scores.

## __g. Sales Performance by Price Range__
Analyze how different price ranges affect sales.

```python
data2 = data.copy()
```

```python
price_bins = [0, 50, 100, 200, 500, 1000]
data2['price_range'] = pd.cut(data2['price'], bins=price_bins)
```

```python
sales_by_price_range = data2.groupby('price_range', observed=False)['sales_month_1'].sum()
sales_by_price_range
```

```python
plt.figure(figsize=(10, 6))
sales_by_price_range.plot(kind='bar', color='lightgreen', edgecolor='k')
plt.title('Sales Performance by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

* Products priced between 200 to 500 have the highest sales volume.
* Lower-priced items (under 50), (under 50-100) and mid-range products (100 to 200) also perform well.
* High-priced items (over 500) has lower sales volumes, suggesting a more selective customer base or niche market.
